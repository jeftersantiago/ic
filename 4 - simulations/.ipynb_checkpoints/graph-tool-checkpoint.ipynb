```python
import graph_tool.all as gt
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Number of nodes in the graph
n = 500

# List of probabilities p for edge creation
pk = np.linspace(0, 0.025, n)

# List with sizes of largest components
largest_cc = []

# Create a graph for each probability p
for p in pk:
    # Create a random graph with probability p
    # er = gt.random_graph(n, lambda: random.random() < p[i], directed=False)
    er = gt.random_graph(n, lambda: np.random.poisson((n-1) * p), directed=False, model="erdos")

    # https://graph-tool.skewed.de/static/doc/autosummary/graph_tool.topology.label_largest_component.html#graph_tool.topology.label_largest_component
    l = gt.label_largest_component(er, directed=None)
    u = gt.GraphView(er, vfilt=l) 
    largest_cc.append(u.num_vertices())
```

```python
n = 500
g = gt.random_graph(n, lambda: np.random.poisson((n-1) * 0.3), directed=False, model="erdos")

# Calculate shortest paths
shortest_paths = gt.shortest_distance(g)

M = []

for i in g.vertices():
    M.append(np.array(shortest_paths[i]))
M = np.array(M)

print(np.mean(M))

# Calculate shortest path lengths
distances = gt.shortest_distance(g)

# Calculate average shortest path length
total_distance = sum(sum(row) for row in distances)
num_pairs = len(g.get_vertices()) * (len(g.get_vertices()) - 1)
avg_distance = total_distance / num_pairs
print("Average shortest path length:", avg_distance)
```

```python
plt.figure(figsize=(10,6))
plt.plot(pk, largest_cc)
plt.title("Percolação em redes de Erdős-Rényi usando graph-tool", fontsize = 16)
plt.xlabel("Probabilidade $p$", fontsize = 16)
plt.ylabel("Tamanho do maior componente", fontsize = 16)
plt.show()
```

## Efeito de mundo pequeno (Strogatz-Watts)

```python
# Gera uma rede de mundo pequeno 
from itertools import chain

def get_n_levels_neighbours(maxn,n,k):
    # l = range(n-int(k/2),n)
    # l += range(n+1,n+int(k/2)+1)

    l = list(chain(range(n-int(k/2),n), range(n+1,n+int(k/2)+1)))

    for i in range(len(l)):
        if (l[i]<0):
            l[i]=((l[i])%(maxn-1))+1
        if(l[i]>(maxn-1)):
            l[i]=l[i]%maxn
    return l
    
def watts_strogatz_network(n,k,p):
    g=gt.lattice([n])
    # make a ring
    g.add_edge(g.vertex(0), g.vertex(n-1))
    # add edges to k neighbours of each vertex in the ring
    # k-1 neighbours if k is odd
    if ((k%2!=0) and k>1):
        k-=1
    if k>2:
        for v in range(n):
            l_n=get_n_levels_neighbours(n,v,k)
            for v_n in l_n:
                if not g.edge(v_n,v):
                    g.add_edge(g.vertex(v), g.vertex(v_n))
    # replace each edge u-v by an edge u-w with probability p
    for u in range(n):
        for v in g.vertex(u).all_neighbours():
            if (random.random()<=p):
                l1=range(n)
                l2=get_n_levels_neighbours(n,u,k)
                l=[i for i in l1 if i not in l2]
                l.remove(u)
                #print l
                w=random.choice(l)
                while w==u or g.edge(u,w):
                    w=random.choice(l)
                g.remove_edge(g.edge(u, v))
                g.add_edge(g.vertex(u), g.vertex(w))
    return g
```

```python
def avg_path_length(g):
    shortest_paths = gt.shortest_distance(g)

    sum_path_lengths = 0
    num_paths = 0
    for v in g.vertices():
        for u in g.vertices():
            if v != u:
                path_length = shortest_paths[v][u]
                if path_length != float('inf'):
                    sum_path_lengths += path_length
                    num_paths += 1

    # Calculate the average path length
    path_length = sum_path_lengths / num_paths
    return path_length
```

```python
n = 500
k = 12

vL = []
vC = []
vP = np.arange(0, 1, 0.01)
for p in vP:
    g = watts_strogatz_network(n, k, p)
    C = np.mean(np.array(gt.global_clustering(g)))
    L = avg_path_length(g)
    vC.append(C)
    vL.append(L)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

plt.plot(vP, np.array(vL)/vL[0], color="red", linewidth=1.3, label = "shortest path length")
plt.plot(vP, np.array(vC)/vC[0], color="blue", linewidth=1.3, label = "clustering")

plt.title("Efeito de mundo pequeno em grafos de Watts-Strogatz usando NetworkX", fontsize = 16)

plt.xlabel("q", fontsize=20)
plt.ylabel("$cc/cc_0$, $\ell/\ell_0$", fontsize=20)

plt.legend()
plt.show()
```
