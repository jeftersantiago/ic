## Bibliotecas utilizadas

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import time
```

```python
def degree_distribution(G):
  vk = np.array(list(dict(G.degree()).values()))
  max_vk = np.max(vk)
  min_vk = np.min(min)

  # Creating a distribution 
  kvalues = np.arange(0, max_vk+1)
  Pk = np.zeros(max_vk+1)
  for i in vk:
    Pk[i] += 1
  Pk = Pk/sum(Pk)
  return kvalues, Pk
```

## Percolação em redes complexas (Erdos-Renyi)



```python
# numero de nós no grafo
n = 500 
# Lista de probabilidades p de criação de arestas.
pk = np.linspace(0, 0.025, n)

# Lista com tamanho dos maiores componentes
largest_cc = []
# Cria grafo de Erdos-Renyi para cada probabilidade p e
# armazena o valor do maior componente na lista.
for p in pk:
    # Lista de grafos aleatorios com probabilidade p.
    er = nx.gnp_random_graph(n,p)
    # Tamanho do maior componente para cada probabilidade.
    # print("components = ", len(list(nx.connected_components(er))))
    largest_cc.append(len(max(nx.connected_components(er), key=len)))
```

```python
n = 500
g = nx.gnp_random_graph(n, 0.3)

print(nx.average_shortest_path_length(g))
```

```python
plt.figure(figsize=(10,6))
plt.plot(pk, largest_cc)
plt.title(r'Percolação em redes de Erdős-Rényi usando NetworkX', fontsize = 16)
plt.xlabel("Probabilidade $p$", fontsize = 16)
plt.ylabel("Tamanho do maior componente", fontsize = 16)
plt.savefig("resultado1.eps")
plt.show()
```

```python
N = 100
S = []
vk = []
for c in np.arange(0, 5, 0.05):
    p = c/(N-1)
    G = nx.fast_gnp_random_graph(N, p, seed=None, directed=False)
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G = G.subgraph(Gcc[0])
    S.append(len(G)) # /N)
    vk.append(c)
    
plt.figure(figsize=(8,6))
plt.plot(vk,S,'b')
plt.xlabel(r'$\langle k \rangle$', fontsize=20)
plt.ylabel("Tamanho do maior componente", fontsize=20)
plt.title(r'Transição de fase em relação ao grau $\langle k \rangle$', fontsize=20)
# plt.axvline(x=1, color = 'r')
plt.show();
```

```python
N = 100
av_degree = 8
p = av_degree/(N-1)
GER = nx.gnp_random_graph(N, p, seed=None, directed=False)

ks, Pk = degree_distribution(GER)


plt.figure(figsize=(8,6))
plt.plot(ks,Pk,'bo', label='Data')
plt.xlabel("k", fontsize=20)
plt.ylabel("P(k)", fontsize=20)
plt.title("Degree distribution", fontsize=20)
plt.grid(True)
# plt.savefig('degree_dist.eps') #save the figure into a file

## Fitting of. binomial distribution
from scipy.stats import binom
pk = binom.pmf(ks, N, p)
plt.plot(ks, pk, 'r', label='Binomial distribution')
plt.legend()
plt.show(True)
```

## Efeito de mundo pequeno (Strogatz-Watts)

```python
n = 500
k = 12

vL = []
vC = []
vP = np.arange(0, 1, 0.001)

for p in vP:
    g = nx.watts_strogatz_graph(n, k, p, seed=None)
    C =  nx.average_clustering(g)
    L = nx.average_shortest_path_length(g)
    vC.append(C)
    vL.append(L)

fig, ax = plt.subplots(figsize=(10, 8))

plt.plot(vP, np.array(vL)/vL[0], color="red", linewidth=1.3, label = "shortest path length")
plt.plot(vP, np.array(vC)/vC[0], color="blue", linewidth=1.3, label = "clustering")

plt.title("Efeito de mundo pequeno em grafos de Watts-Strogatz usando NetworkX", fontsize = 16)

plt.xlabel("q", fontsize=20)
plt.ylabel("$cc/cc_0$, $\ell/\ell_0$", fontsize=20)
plt.savefig("resultado2.eps")
plt.legend()
plt.show()
```

## Resiliência a ataques aleatórios e direcionados (Albert, Barabási e Jeong)

```python
def failures(H):
    G = H.copy()
    from random import choice
    N0 = len(G)
    vn = []
    S = []
    n = 0
    while(len(G.nodes()) > int(0.01*N0)):
        node = random.choice(list(G.nodes()))
        G.remove_node(node) 
        #Gcc=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)
        Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
        Glc = G.subgraph(Gcc[0])        
        S.append(len(Glc)/N0) 
        n = n + 1
        vn.append(n/N0)
    return S, vn
```

```python
# encontra o vertice mais conectado do grafo 
def most_connected(G): 
    maxk = 0
    node = 0
    for i in G.nodes():
        if(G.degree(i) >= maxk):
            maxk = G.degree(i)
            node = i
    return node 

def attacks(H):
    G = H.copy()
    from random import choice
    N0 = len(G)
    vn = []
    S = []
    # numero de vertices removidos
    n = 0 
    while(len(G.nodes()) > int(0.01*N0)):
        node = most_connected(G)             
        G.remove_node(node) 
        Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
        Glc = G.subgraph(Gcc[0])
        S.append(len(Glc)/N0)
        n = n + 1
        vn.append(n/N0)
    return S, vn
```

```python
N = 500
av_degree = 10
p = av_degree/(N-1)
# GER = nx.gnp_random_graph(N, p, seed=None, directed=False)
m = int(av_degree/2)
g = nx.barabasi_albert_graph(N, m)

N = len(g)
M = g.number_of_edges()
print('Numero de vertices:', N)
print('Numero de arestas:', M)

Sf, vnf = failures(g)
Sa, vna = attacks(g)

plt.figure(figsize=(12, 12))
plt.figure()
plt.title('Resiliência a ataques aleatórios em \n redes de Barabási-Albert  usando NetworkX', fontsize=15)
plt.plot(vnf,Sf,'-or', label='Falhas')
plt.plot(vna,Sa,'-ob', label='Ataques')
plt.xlabel("f", fontsize=20)
plt.ylabel("S", fontsize=20)
# plt.axvline(x=fc, color = 'g', label='fc failures')
plt.legend()
plt.grid(True)
plt.savefig("resultado3.eps")
plt.show(); 
```
