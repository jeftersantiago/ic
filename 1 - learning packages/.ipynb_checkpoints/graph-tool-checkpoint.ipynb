## Grafos ponderados   

```python
g = gt.Graph(directed = False)
edgelist = [(0, 1, 0.1), (1, 2, 0.5), (0, 2, 0.9), (2, 3, 0.5), (2, 4, 0.2), (1, 3, 0.5), (4, 0, 0.9), (5, 2, 0.6),
           (5, 6, 0.1), (5, 7, 0.8), (5, 8, 0.9), (1, 9, 0.6), (9, 6, 0.9)]

eweight = g.new_ep("double")
g.add_edge_list(edgelist, eprops=[eweight])

pos = gt.sfdp_layout(g)
gt.graph_draw(g, vertex_text=g.vertex_index, output_size=(300, 300))
```

- Desenhando grafo com arestas de largura relativa ao peso.

```python
ebet = gt.betweenness(g)[1]
ebet.a /= ebet.a.max() / 10.
pos = gt.sfdp_layout(g)
gt.graph_draw(g, pos=pos, vertex_text=g.vertex_index, edge_pen_width=ebet, output_size=(300, 300))
```

## Grafo de Erdos-Renyi (Modelo $G_{np}$)   

Os modelos especificos de grafos aleatórios foram bem complicados de fazer, na minha opinião. Na documentação da biblioteca 
https://graph-tool.skewed.de/static/doc/generation.html informa como cria um grafo aleatório e a forma de escolher 
o modelo consiste em gerar uma função que cria as ``amostras'', ou seja, tem que descrever o tipo de distribuição que segue o modelo. Isso pode ser prático pra determinados modelos cuja distribuição do grau é mais simples, mas por exemplo para o modelo do mundo pequeno (Watts-Strogatz) fica inviavel gerar os grafos dessa forma. Nesse caso não consegui encontrar na mesma biblioteca uma forma de gerar um grafo de Watts-Strogatz e então fiz uma implementação manualmente.

```python
n = 100
k = 3 
p = k / (n - 1)

g = gt.random_graph(n, lambda: np.random.poisson((n-1) * p), directed=False, model="erdos")

pos = gt.arf_layout(g)
gt.graph_draw(g, pos=pos) 
print("Number of nodes:", g.num_vertices())
print("Number of edges:", g.num_edges())

total_degree = sum(g.vertex(v).out_degree() for v in g.vertices())
mean_degree = total_degree / g.num_vertices()
print("<k> = ", mean_degree)
```

## Grafo modelo Watts-Strogatz - Rede de mundo pequeno

```python
from itertools import chain

def get_n_levels_neighbours(maxn,n,k):
    # l = range(n-int(k/2),n)
    # l += range(n+1,n+int(k/2)+1)

    l = list(chain(range(n-int(k/2),n), range(n+1,n+int(k/2)+1)))

    for i in range(len(l)):
        if (l[i]<0):
            l[i]=((l[i])%(maxn-1))+1
        if(l[i]>(maxn-1)):
            l[i]=l[i]%maxn
    return l
    
def watts_strogatz_network(n,k,p):
    g=gt.lattice([n])
    # make a ring
    g.add_edge(g.vertex(0), g.vertex(n-1))
    # add edges to k neighbours of each vertex in the ring
    # k-1 neighbours if k is odd
    if ((k%2!=0) and k>1):
        k-=1
    if k>2:
        for v in range(n):
            l_n=get_n_levels_neighbours(n,v,k)
            for v_n in l_n:
                if not g.edge(v_n,v):
                    g.add_edge(g.vertex(v), g.vertex(v_n))
    # replace each edge u-v by an edge u-w with probability p
    for u in range(n):
        for v in g.vertex(u).all_neighbours():
            if (random.random()<=p):
                l1=range(n)
                l2=get_n_levels_neighbours(n,u,k)
                l=[i for i in l1 if i not in l2]
                l.remove(u)
                #print l
                w=random.choice(l)
                while w==u or g.edge(u,w):
                    w=random.choice(l)
                g.remove_edge(g.edge(u, v))
                g.add_edge(g.vertex(u), g.vertex(w))
    return g


n = 50
k = 6
p = 0.3

g = watts_strogatz_network(n, k, p)
```
