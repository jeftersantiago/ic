# Geração de grafos aleatórios

## Modelo Erdos-Renyi


- A biblioteca `graph-tool` é dificil de usar. Não consegui de forma alguma usar as 
funções que eles fornecem para geração de grafos aleatórios seguindo os modelos mais padrão, como Erdos-Renyi e Barabasi-Albert. Por causa disso fiz as implementações manualmente. Essa opção que tive influi na qualidade desses códigos, já que não é escrito pela própria biblioteca. 
- A dificuldade de usar a `graph-tool` é, ao meu ver, culpa da documentação, que possui poucos exemplos e qunado tem exemplos não são de casos mais básicos, como os que pessoas que estão tentando aprender a usa-lá vão precisar.
- Poderia-se dizer que é um maior desafio usar a `graph-tool` por conta da sua maior integração com `C++` e por não usar o paradigma de orientação à objetos. No entanto, não acredito que seja o caso, pois essas dificuldades não se mostram no uso da `igraph`, que é construída de forma bem semelhante.

```python
Ns = [100, 200, 300, 400, 500, 600, 700]
Pk = np.array(np.random.random(len(Ns)))

N = len(Ns)


vIG = []
vGT = []
vNX = []

for k in range(0, N):
    vIG_time = []
    vGT_time = []
    vNX_time = []
    for p, n in zip(Pk, Ns):
        t0 = time.time()
        ger_nx = nx.gnp_random_graph(n, p, seed=None)
        vNX_time.append(time.time() - t0)
       
       # N = len(ger_nx)
       # M = ger_nx.number_of_edges()
       # print('Numero de vertices:', N)
       # print('Numero de arestas:', M)        

        t0 = time.time()
        ger_ig = ig.Graph.Erdos_Renyi(n, p)
        vIG_time.append(time.time() - t0)

        t0 = time.time()
        ger_gt = gt.random_graph(n, lambda: np.random.poisson((n-1) * p), directed=False, model="erdos")
        vGT_time.append(time.time() - t0)
        
        
    vIG.append(vIG_time)
    vGT.append(vGT_time)
    vNX.append(vNX_time)


vIG_median = np.median(np.array(vIG), axis = 0)
vGT_median = np.median(np.array(vGT), axis = 0)
vNX_median = np.median(np.array(vNX), axis = 0)


plt.figure(figsize=(8, 8))

plt.plot(Ns, vIG_median,'o', label = "igraph", color = "red")
plt.plot(Ns, vNX_median,'o',  label = "NetworkX", color = "blue")
plt.plot(Ns, vGT_median, 'o',  label = "graph-tool", color = "gray")

plt.legend(fontsize = 20)
plt.ylabel("Tempo em segundos", fontsize = 20)
plt.xlabel("Número de vértices", fontsize = 20)
plt.title('Tempo de geração de redes no modelo \n de Erdős-Renyi, em função do número de vértices', fontsize = 20)
plt.savefig("comparacao_gerador_grafo.eps")
plt.show()
```

# Medidas locais em redes

## Betweenness centrality

```python
# N = len(network_paths)

vIG = []
vGT = []
vNX = []
Ns = []
Ns = [len(nx.read_edgelist(network)) for network in network_paths]

for i in range(0, N):
    vIG_time = []
    vGT_time = []
    vNX_time = []
    
    # print(len(G_nx))
    for network in network_paths:
        G_nx = nx.read_edgelist(network)
        
        t0 = time.time()
        nx.betweenness_centrality(G_nx)
        vNX_time.append(time.time() - t0)

        G_ig = ig.Graph.Read_Edgelist(network)
        t0 = time.time()
        ig_betweenness = G_ig.betweenness()
        vIG_time.append(time.time() - t0)

        G_gt = gt.load_graph_from_csv(network, directed=False, csv_options={'delimiter': ' '})
        t0 = time.time()
        gt_betweenness = gt.betweenness(G_gt)[0]
        vGT_time.append(time.time() - t0)

    vIG.append(vIG_time)
    vGT.append(vGT_time)
    vNX.append(vNX_time)


vIG_median = np.median(np.array(vIG), axis = 0)
vGT_median = np.median(np.array(vGT), axis = 0)
vNX_median = np.median(np.array(vNX), axis = 0)

vIG = np.array(vIG_median)
vGT = np.array(vGT_median)
vNX = np.array(vNX_median)
```

```python
plt.figure(figsize=(8, 8))

plt.plot(Ns, vIG_median,'o', label = "igraph", color = "red")
plt.plot(Ns, vNX_median,'o',  label = "NetworkX", color = "blue")
plt.plot(Ns, vGT_median, 'o',  label = "graph-tool", color = "gray")

plt.legend(fontsize = 20)
plt.ylabel("Tempo em segundos", fontsize = 20)
plt.xlabel("Número de vértices", fontsize = 20)
plt.title('Execução da medida betweenness centrality \n em função do número de vértices', fontsize = 20)
plt.savefig("betweenness_centrality.eps")
plt.show()
```

## Closeness centrality

```python
# N = len(network_paths)

vIG = []
vGT = []
vNX = []
Ns = []
Ns = [len(nx.read_edgelist(network)) for network in network_paths]

for i in range(0, N):
    vIG_time = []
    vGT_time = []
    vNX_time = []
    
    # print(len(G_nx))
    for network in network_paths:
        G_nx = nx.read_edgelist(network)
        
        t0 = time.time()
        avg_shortest_path_length = nx.average_shortest_path_length(G_nx)
        vNX_time.append(time.time() - t0)

        G_ig = ig.Graph.Read_Edgelist(network)
        t0 = time.time()
        closeness = G_ig.average_path_length()
        vIG_time.append(time.time() - t0)

        G_gt = gt.load_graph_from_csv(network, directed=False, csv_options={'delimiter': ' '})
        t0 = time.time()

        total_dist = sum(sum(row) for row in gt.shortest_distance(G_gt))
        num_pairs = len(G_gt.get_vertices()) * (len(G_gt.get_vertices()) - 1)
        avg_shortest_dist = total_dist / num_pairs
        
        vGT_time.append(time.time() - t0)

    vIG.append(vIG_time)
    vGT.append(vGT_time)
    vNX.append(vNX_time)


vIG_median = np.median(np.array(vIG), axis = 0)
vGT_median = np.median(np.array(vGT), axis = 0)
vNX_median = np.median(np.array(vNX), axis = 0)

vIG = np.array(vIG_median)
vGT = np.array(vGT_median)
vNX = np.array(vNX_median)



plt.figure(figsize=(8, 8))

plt.plot(Ns, vIG_median,'o', label = "igraph", color = "red")
plt.plot(Ns, vNX_median,'o',  label = "NetworkX", color = "blue")
plt.plot(Ns, vGT_median, 'o',  label = "graph-tool", color = "gray")

plt.legend(fontsize = 20)
plt.ylabel("Tempo em segundos", fontsize = 20)
plt.xlabel("Número de vértices", fontsize = 20)
plt.title('Execução da medida média de comprimento \n de menor distância \n em função do número de vértices', fontsize = 20)
plt.savefig("avg_shortest_path_length.eps")
plt.show()
```

# Detecção de comunidades - Método Louvain

```python
from networkx.algorithms.community import greedy_modularity_communities


# N = len(network_paths)

vIG = []
vGT = []
vNX = []
Ns = []
Ns = [len(nx.read_edgelist(network)) for network in network_paths]

for i in range(0, N):
    vIG_time = []
    vGT_time = []
    vNX_time = []
    
    # print(len(G_nx))
    for network in network_paths:
        G_nx = nx.read_edgelist(network)
        
        t0 = time.time()

        partitions = greedy_modularity_communities(G_nx)

        
        vNX_time.append(time.time() - t0)

        G_ig = ig.Graph.Read_Edgelist(network, directed = False)
        t0 = time.time()

        louvain_clusters = G_ig.community_multilevel()

        
        vIG_time.append(time.time() - t0)

        G_gt = gt.load_graph_from_csv(network, directed=False, csv_options={'delimiter': ' '})
        t0 = time.time()

        state = gt.minimize_nested_blockmodel_dl(G_gt)

            
        vGT_time.append(time.time() - t0)

    vIG.append(vIG_time)
    vGT.append(vGT_time)
    vNX.append(vNX_time)


vIG_median = np.median(np.array(vIG), axis = 0)
vGT_median = np.median(np.array(vGT), axis = 0)
vNX_median = np.median(np.array(vNX), axis = 0)

vIG = np.array(vIG_median)
vGT = np.array(vGT_median)
vNX = np.array(vNX_median)

plt.figure(figsize=(8, 8))

plt.plot(Ns, vIG_median,'o', label = "igraph", color = "red")
plt.plot(Ns, vNX_median,'o',  label = "NetworkX", color = "blue")
plt.plot(Ns, vGT_median, 'o',  label = "graph-tool", color = "gray")

plt.legend(fontsize = 20)
plt.ylabel("Tempo em segundos", fontsize = 20)
plt.xlabel("Número de vértices", fontsize = 20)
plt.title('Detecção de comunidades pelo método Louvain \n em função do número de vértices', fontsize = 20)
plt.savefig("comunidades.eps")
plt.show()
```
